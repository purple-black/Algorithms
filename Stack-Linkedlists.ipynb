```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class StackList:
    
    def __init__(self):
        self.top = None
        
    
    def create(self):
        size = int(input("how many elements in stack: "))
        for i in range(size):
            val = int(input("enter value to be pushed: "))
            new =  Node(val)
            new.next = self.top
            self.top = new
            
    def display(self):
        internode = self.top
        if self.top == None:
            print("---stack underflow---")
        print("stack----\n")
        while internode is not None:
            print(internode.data)
            internode = internode.next
            
    def push(self,val):
        new = Node(val)
        new.next = self.top
        self.top = new
        
    def pop(self):
        if self.top == None:
            print("---stack empty---")
            return
        ptr = self.top
        self.top = self.top.next
        del(ptr)
        
mystack = StackList()
mystack.create()

while True:
    choice = int(input("select operation: \n1.Push\n2.pop\n3.display\n4.exit\n"))
    if choice == 1:
        item = int(input("enter value to be pushed: "))
        mystack.push(item)
    elif choice == 2:
        mystack.pop()
    elif choice == 3:
        mystack.display()
    else:
        exit()
        
        
            
        
```
