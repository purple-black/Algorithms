```python
class StackArray:
    def __init__(self,size,sz):
        self.arr = [None]*size
        self.maxsize = size
        self.top = -1
        
    def push(self,item):
        if self.isFull():
            print("---Stack Overflow---")
            return
        else:
            self.top = self.top + 1
            self.arr[self.top] = item
            
    def pop(self):
        if self.isEmpty():
            print("---Stack Is Empty---")
            return
        top = self.arr[self.top]
        self.arr[self.top] = None
        self.top = self.top - 1
        return top
       
    
    def size(self):
        return self.top + 1
    
    def isEmpty(self):
        if self.size() == 0:
            return True
        else:
            return False
        
    def isFull(self):
        if self.size() == self.maxsize:
            return True
        else:
            return False
        
    def display(self):
        if self.isEmpty():
            print("---stack empty---")
        else:
            print(self.arr)
        
    
maxsz = int(input("enter maximum size of stack: "))
sz = int(input("enter no of elements in stack: "))
mystack = StackArray(maxsz,sz)
mystack.push(7)
mystack.push(10)
mystack.push(5)

while True:
    choice = int(input("enter the operation you want to perform: \n1. Push\n2. Pop\n3. Display\n. Exit"))
    if choice == 1:
        val = int(input("enter the value to be pushed: "))
        mystack.push(val)
    elif choice == 2:
        p = mystack.pop()
        print(f"popped element: {p}")
    elif choice == 3:
        mystack.display()
    else:
        exit()
            

```
