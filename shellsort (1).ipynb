```python
#insertion sort

def insertionSort(b):
    for i in range(1,len(b)):
        value = b[i]
        hole = i
        
        while hole > 0 and b[hole-1] > value:
            b[hole] = b[hole - 1]
            hole -= 1
            
        b[hole] = value
        
    print("array after insertion sort: ")
    print(b)


b = []
for i in range(5):
    val = random.randint(1,10)
    b.append(val)
    
print("array before insertion sort: ")
print(b)
    
insertionSort(b)
```
